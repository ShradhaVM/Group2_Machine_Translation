```python
import re
from nltk.translate.bleu_score import sentence_bleu
import warnings
warnings.filterwarnings('ignore')
import random
import torch
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print("Device: ", device)
print("GPU:", torch.cuda.get_device_name(torch.cuda.current_device()))
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
tokenizer = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-hi-en")
model = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-hi-en").to(device)
```

```python
def bleu(src,tgt):
  total = 0
  total_i = 0
  for k in range(len(src)//50):
    j = 50*k
    src_ = src[j:j+50]
    inputs = tokenizer(src_, return_tensors="pt",padding=True,max_length=100,truncation=True).to(device)
    outputs = model.generate( inputs.input_ids, max_length=100, num_beams=None, early_stopping=True)
    sen = []
    for out in outputs:
      sen.append(tokenizer.decode(out))
    for i in range(len(sen)):
      sen[i] = re.sub("<pad>","",sen[i].lower())  
    
    running_bleu = 0
    tgt_ = tgt[j:j+50]
    p =0
    for i in range(len(sen)):
      try: 
        running_bleu += sentence_bleu(tgt_[i].split(),sen[i].split())
        p += 1
      except:
        pass  
    r = random.randint(0,49)
    if k%5 == 0:
      print("Hindi Sentence :",src_[r])
      print("English Sentence :",sen[r])
        
    print("Score for batch {0} : {1}".format(k+1,running_bleu/p))
    total += running_bleu
    total_i += 50

  return total/total_i,total
```

```python
bleu_,total = bleu(src,tgt)
```

```python
print("Final Bleu : ",total/500)
```

```python
def translate(hin_snippet):

  hin_snippet = hin_snippet.split('\u0964')
  inputs = tokenizer(hin_snippet, return_tensors="pt",padding=True,max_length=500,truncation=True)
  outputs = model.generate(inputs.input_ids.cuda(), max_length=500, num_beams=None, early_stopping=True)
  sen = []
  for out in outputs:
    sen.append(tokenizer.decode(out))
  for i in range(len(sen)):
    sen[i] = re.sub("<pad>","",sen[i].lower())
  sen = ' '.join(sen)
  return sen
```

```python
from flask import Flask, render_template, url_for, request
from flask_ngrok import run_with_ngrok

app = Flask(__name__)

run_with_ngrok(app)  
 

@app.route('/')
@app.route('/home')

def home():
    return render_template("index.html")


@app.route('/result',methods=['POST', 'GET'])
def result():
    
    output = request.form.to_dict()

    print(output)
    text = output["name"]
    translated_text = translate(text)


    return render_template('index.html', hindi_text=text, name = translated_text)
    
if __name__ == "__main__":
    app.run()
```
